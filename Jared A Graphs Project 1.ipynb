```python
import pandas as pd

import matplotlib.pyplot as plt

cause_death = pd.read_csv('cause_of_deaths.csv')

cause_death.head()
```

```python
cause_death.dropna()
```

```python
for col in cause_death.columns:
    print(col)
```

```python
countries_continents = pd.read_csv('countries_continents.csv')

del countries_continents['Unnamed: 0']
countries_continents
```

```python
for index, row in cause_death.iterrows():

    country = cause_death.loc[index, 'Country/Territory'] 
    #find this country in the count cont file
    
    row_cont = countries_continents.loc[countries_continents['Country'] == country]
    try: 
        cause_death.loc[index, 'Continent'] = row_cont['Continent'].iat[0]
    #leaving out territories for now !
    except IndexError:
        pass
    
    

cause_death
```

```python
continent_year = cause_death[cause_death['Year'] == 1990].groupby('Continent').count()
continent_year
```

```python
for continent in continent_year.index:
    #print(continent)
    cause_continent = cause_death[cause_death['Continent'] == continent]
    year_continent = cause_continent.groupby('Year').sum()
    plt.plot(year_continent.index, year_continent['Interpersonal Violence'])
    
    
plt.legend(continent_year.index)
plt.title('Interpersonal Violence vs Year')
plt.ylabel('Interpersonal Violence Cases')
```

```python
disease_by_continent = pd.read_csv('total-disease-burden-by-cause.csv')

disease_by_continent.head()
```

```python
disease_by_continent.rename(columns = {'Entity':'Country/Territory'})
```

```python
continents_df = pd.merge(cause_death,disease_by_continent,right_index=True, left_index=True)
continents_df
```

```python
continent_df = continents_df[['Year_x','Continent','DALYs (Disability-Adjusted Life Years) - Injuries - Sex: Both - Age: All Ages (Number)','DALYs (Disability-Adjusted Life Years) - Communicable, maternal, neonatal, and nutritional diseases - Sex: Both - Age: All Ages (Number)','DALYs (Disability-Adjusted Life Years) - Non-communicable diseases - Sex: Both - Age: All Ages (Number)']].copy()
continent_df
```

```python
for continent in continent_df.index:
    
    DALYs_continent = continent_df[continent_df['Continent'] == continent] 
    year_continent = continent_df.groupby('Year_x').sum()
    #plt.rcParams["figure.figsize"] = [7.50, 3.50]
    plt.rcParams["figure.autolayout"] = True
    #lt.ylim(0, 10)
    plt.plot(year_continent.index, year_continent['DALYs (Disability-Adjusted Life Years) - Injuries - Sex: Both - Age: All Ages (Number)'])
    
    
plt.legend(continent_year.index)
plt.title('Injuries vs Year')
plt.ylabel('Injuries by Continent Cases')

```
